```sql
USE CATALOG workspace;
USE SCHEMA portfolio_jobs;

CREATE OR REPLACE VIEW v_gold_post_trend AS
SELECT posted_date, job_posts
FROM gold_post_trend;

CREATE OR REPLACE VIEW workspace.portfolio_jobs.v_gold_top_companies AS
WITH ranked AS (
  SELECT
    company_norm AS company,
    CAST(job_posts AS BIGINT) AS job_posts,
    ROW_NUMBER() OVER (
      ORDER BY CAST(job_posts AS BIGINT) DESC, company_norm ASC
    ) AS rn
  FROM workspace.portfolio_jobs.gold_top_companies
)
SELECT company, job_posts
FROM ranked
WHERE rn <= 20
ORDER BY rn;
```
