```python
from pyspark.sql import functions as F


# Access the raw dataset:
catalog = "bright_data_linkedin_jobs_listing_dataset"
schema  = "datasets"
table   = "linkedin_full_job_listings_information"  # from SHOW TABLES

src = spark.table(f"{catalog}.{schema}.{table}")
display(src.limit(20))
src.printSchema()
print("rows =", src.count())

# Null rates for a few key columns (adjust after schema inspection)
cols = ["job_title", "company_name", "job_location", "job_posted_time", "url"]
for c in [x for x in cols if x in src.columns]:
    src.select(
        F.lit(c).alias("col"),
        F.count(F.lit(1)).alias("rows"),
        F.sum(F.col(c).isNull().cast("int")).alias("nulls")
    ).show(truncate=False)
```
