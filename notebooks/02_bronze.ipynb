```python
from pyspark.sql import functions as F


catalog = "bright_data_linkedin_jobs_listing_dataset"
schema  = "datasets"
table   = "linkedin_full_job_listings_information"

src = spark.table(f"{catalog}.{schema}.{table}")
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS portfolio_jobs;
```

```python

bronze = (src
  .withColumn("_ingest_ts", F.current_timestamp())
  .withColumn("_source", F.lit("bright_data_linkedin_jobs_listing_dataset"))
)

(bronze.write
  .mode("overwrite")
  .format("delta")
  .saveAsTable("portfolio_jobs.bronze_linkedin_jobs"))
```
